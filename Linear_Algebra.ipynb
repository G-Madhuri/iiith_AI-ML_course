
**Inner Product:**
The inner product is the scalar multiplication of one vector (or matrix) and the transpose of another vector (or matrix). If both arrays are 1D arrays, their dimensions should be identical. If either or both arrays are higher-dimensional, then the last dimensions of both arrays should be identical.

```python
a = np.array([[4,5,6], [1,2,3]])
b = np.array([7,8,9])
z = np.inner(a,b)
print(z)
a.dot(b.T)
```

**Outer Product:**
Outer product is the dot product of a column vector of size Mx1 and a row vector of size 1xN. The resulting array is a matrix of size MxN.

```python
a = np.array([1,2,3,4,5])
b = np.array([6,3,2])
z = np.outer(a,b)
print(z)
```

**Matrix Factorization**

```python
A = np.array([[12, -51, 4],
              [6, 167, -68],
              [-4, 24, -41],
              [-1, 1, 0]])
Q, R = np.linalg.qr(A)

print("Matrix A:")
print(A)
print("\nMatrix Q (Orthogonal):")
print(Q)
print("\nMatrix R (Upper Triangular):")
print(R)

print("\nVerification (A = QR):")
print(np.allclose(A, Q @ R))
```

**Singular Value Decomposition**

```python
from skimage import data
image = data.cat()
print(image.dtype, image.min(), image.max(), image.shape)
plt.imshow(image)
plt.show()
```

```python
img = image/255.0
img = np.transpose(img, axes=(2,0,1))
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
```

```python
Sigma = np.zeros((3,300,451))
for i in range(3):
  np.fill_diagonal(Sigma[i,:,:], S[i,:])
print(Sigma.shape)
reconst = U @ Sigma @ V
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

```python
k = 25
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```

**Dimensionality reduction**

```python
k = 10
reconst = U @ Sigma[:,:,:k] @ V[:,:k,:]
reconst = np.transpose(reconst, axes=(1,2,0))
plt.imshow(reconst)
plt.show()
```
