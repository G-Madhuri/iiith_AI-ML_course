## Gradient Descent

```python
##########################
##########################

## TASK-1 : Complete the lines of code wherever marked as [REQUIRED] in this cell.

##########################
##########################

# Gradient Descent
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.00001

x_list = [new_x]
slope_list = [df(new_x)]

for n in range(500):
    previous_x = new_x
    gradient = df(previous_x)  # [REQUIRED]
    new_x = previous_x - step_multiplier * gradient  # [REQUIRED]

    step_size = abs(new_x - previous_x)
    x_list.append(new_x)
    slope_list.append(df(new_x))  # [REQUIRED]

    if step_size < precision:
        print('Loop ran this many times:', n)
        break

print('Local minimum occurs at:', new_x)
print('Slope or df(x) value at this point is:', df(new_x))
print('f(x) value or cost at this point is:', f(new_x))
```

## Gradient Descent

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001,
                    max_iter=300):
    new_x = initial_guess
    x_list = [new_x]
    slope_list = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))

        if step_size < precision:
            break
    return new_x, x_list, slope_list
```

## Minimise $$f(x, y) = \frac{1}{3^{-x^2 - y^2} + 1}$$

Minimise $$f(x, y) = \frac{1}{r + 1}$$ where $r$ is $3^{-x^2 - y^2}$

```python
## TASK-8 : Code the same things as they were done for the first two examples
# Define the 2D function
f = lambda x, y: (1 / (np.exp(-x**2 - y**2) + 1))

def df_dx(x, y):
    return (2 * x * np.exp(-x**2 - y**2)) / (np.exp(-x**2 - y**2) + 1)**2

def df_dy(x, y):
    return (2 * y * np.exp(-x**2 - y**2)) / (np.exp(-x**2 - y**2) + 1)**2

# Gradient descent for 2D functions
def gradient_descent_2d(f_dx, f_dy, initial_guess, multiplier=0.1, precision=0.0001, max_iter=500):
    x, y = initial_guess
    x_list, y_list = [x], [y]

    for n in range(max_iter):
        grad_x, grad_y = f_dx(x, y), f_dy(x, y)
        x_new = x - multiplier * grad_x
        y_new = y - multiplier * grad_y

        if abs(x_new - x) < precision and abs(y_new - y) < precision:
            print(f"Gradient descent converged after {n} iterations.")
            break

        x, y = x_new, y_new
        x_list.append(x)
        y_list.append(y)

    return (x, y), x_list, y_list

# Running gradient descent for TASK-8
initial_guess = (1.5, -1.5)
local_min, x_vals, y_vals = gradient_descent_2d(df_dx, df_dy, initial_guess, multiplier=0.1)
print("Local minimum occurs at:", local_min)
print("Function value at local minimum:", f(local_min[0], local_min[1]))

# Plotting the 2D function and gradient descent path
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
x, y = np.meshgrid(x, y)
z = f(x, y)

fig = plt.figure(figsize=(15, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap=cm.coolwarm, alpha=0.8)
ax.set_title("Surface Plot of f(x, y)")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

# Plot gradient descent path
ax.scatter(x_vals, y_vals, [f(x, y) for x, y in zip(x_vals, y_vals)], color='red', label="Gradient Descent Path")
plt.legend()
plt.show()
```

## Partial Derivatives & Symbolic Computation

## $$\frac{\partial f}{\partial x} = \frac{2x e^{-x^2 - y^2}}{(e^{-x^2 - y^2} + 1)^2}$$

## $$\frac{\partial f}{\partial y} = \frac{2y e^{-x^2 - y^2}}{(e^{-x^2 - y^2} + 1)^2}$$

```python
from sympy import symbols, exp

# Define the symbolic cost function
a, b = symbols('x y')
f_sym = 1 / (exp(-a**2 - b**2) + 1)

# Compute symbolic partial derivatives
df_dx_sym = diff(f_sym, a)
df_dy_sym = diff(f_sym, b)

# Output the results
print('Our cost function f(x, y) is: ', f_sym)
print('Partial derivative wrt x is: ', df_dx_sym)
print('Partial derivative wrt y is: ', df_dy_sym)
f_value = f_sym.evalf(subs={a: 1.8, b: 1.0})
print('Value of f(x, y) at x=1.8 and y=1.0:', f_value)
```

## Batch Gradient Descent with SymPy

```python
from sympy import lambdify

f_numeric = lambdify((a, b), f_sym, modules='numpy')

# Setup
multiplier = 0.1
max_iter = 500
params = np.array([1.8, 1.0])  # Initial guess

for n in range(max_iter):
    gradient_x = diff(f_sym, a).evalf(subs={a: params[0], b: params[1]})
    gradient_y = diff(f_sym, b).evalf(subs={a: params[0], b: params[1]})
    gradients = np.array([gradient_x, gradient_y], dtype=float)
    params = params - multiplier * gradients

# Results
cost = f_numeric(params[0], params[1])
print('Values in gradient array:', gradients)
print('Minimum occurs at x value of:', params[0])
print('Minimum occurs at y value of:', params[1])
print('The cost is:', cost)
```

## Mean Squared Error: a cost function for regression problems

### $$RSS = \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y^{(i)} - h_\theta x^{(i)} \big)^2 $$
### $$MSE = \frac{1}{n} \sum_{i=1}^{n} \big( y - \hat{y} \big)^2 $$

```python
# Make sample data
x_5 = np.array([[0.1, 1.2, 2.4, 3.2, 4.1, 5.7, 6.5]]).transpose()
y_5 = np.array([1.7, 2.4, 3.5, 3.0, 6.1, 9.4, 8.2]).reshape(7, 1)

print('Shape of x_5 array:', x_5.shape)
print('Shape of y_5 array:', y_5.shape)
```

```python
# Quick linear regressino
regr = LinearRegression()
regr.fit(x_5, y_5)
print('Theta 0:', regr.intercept_[0])
print('Theta 1:', regr.coef_[0][0])
```

```python
# Challenge: Write a python function mse(y, y_hat) returns the MSE? Call the mse(y, y_hat)
# function and print out the MSE for the y_hat calculated above.


def mse(y, y_hat):
    #mse_calc = 1/7 * sum((y - y_hat)**2)
    #mse_calc = (1/y.size) * sum((y - y_hat)**2)
    mse_calc = np.average((y - y_hat)**2, axis=0)
    return mse_calc
```

## Partial Derivatives of MSE w.r.t. $\theta_0$ and $\theta_1$

## $$\frac{\partial MSE}{\partial \theta_0} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big)$$

## $$\frac{\partial MSE}{\partial \theta_1} = - \frac{2}{n} \sum_{i=1}^{n} \big( y^{(i)} - \theta_0 - \theta_1 x^{(i)} \big) \big( x^{(i)} \big)$$

## MSE & Gradient Descent

```python
# x values, y values, array of theta parameters (theta0 at index 0 and theta1 at index 1)
def grad(x, y, thetas):
    n = y.size

    # Challenge: Create theta0_slope and theta1_slope to hold slope values from partial derivs
    theta0_slope = (-2/n) * sum(y - thetas[0] - thetas[1]*x)
    theta1_slope = (-2/n) * sum((y - thetas[0] - thetas[1]*x)*x)

    #return np.array([theta0_slope[0], theta1_slope[0]])
    #return np.append(arr=theta0_slope, values=theta1_slope)
    return np.concatenate((theta0_slope, theta1_slope), axis=0)
```

```python
multiplier = 0.01
thetas = np.array([2.9, 2.9])

# Collect data points for scatter plot
plot_vals = thetas.reshape(1, 2)
mse_vals = mse(y_5, thetas[0] + thetas[1]*x_5)

for i in range(1000):
    thetas = thetas - multiplier * grad(x_5, y_5, thetas)

    # Append the new values to our numpy arrays
    plot_vals = np.concatenate((plot_vals, thetas.reshape(1, 2)), axis=0)
    mse_vals = np.append(arr=mse_vals, values=mse(y_5, thetas[0] + thetas[1]*x_5))

# Results
print('Min occurs at Theta 0:', thetas[0])
print('Min occurs at Theta 1:', thetas[1])
print('MSE is:', mse(y_5, thetas[0] + thetas[1]*x_5))
```
